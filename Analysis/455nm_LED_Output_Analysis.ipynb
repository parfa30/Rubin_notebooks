# LED Scan Analysis

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd_response = pd.read_csv('hamamatsu_responsivity.csv')
area = 100 #mm2
amps_per_watt = float(pd_response[pd_response.wavelength == 440].responsivity)
```

```python
def watts_per_mm2(amps):  
    watts_per_mm2 = (amps/amps_per_watt)/area
    return watts_per_mm2
```

```python
data_path = '/home/parfa30/DATA/ScanData'
scan_0mm = os.path.join(data_path,'LED_center_scan_0mm.csv')
scan_2mm = os.path.join(data_path,'LED_center_scan_2mm.csv')
scan_4mm = os.path.join(data_path,'LED_center_scan_4mm.csv')
scan_5mm = os.path.join(data_path,'LED_center_scan_5mm.csv')
scan_6mm = os.path.join(data_path,'LED_center_scan_6mm2.csv')
scan_7mm = os.path.join(data_path,'LED_center_scan_7mm2.csv')
scan_8mm = os.path.join(data_path,'LED_center_scan_8mm2.csv')
scan_9mm = os.path.join(data_path,'LED_center_scan_9mm.csv')
scan_10mm = os.path.join(data_path,'LED_center_scan_10mm.csv')
scan_1 = os.path.join(data_path,'LED_center_scan_20220412.csv')
scan_2 = os.path.join(data_path,'LED_center_scan_loc1_20220413.csv')
scan_fine = os.path.join(data_path, 'LED_center_scan_fine_20220412.csv')
scan_full = os.path.join(data_path, 'LED_center_scan_full_20220412.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,6))
centers = [2150, 2150, 2150, 2180, 2100, 2175, 2125, 2175]
name = [0,2,4,5,6,7,8,9]
for i, df in enumerate([scan_0mm, scan_2mm, scan_4mm, scan_5mm, scan_6mm, scan_7mm, scan_8mm, scan_9mm]):
    data = pd.read_csv(df)
    ax1.plot(data['X']-centers[i], watts_per_mm2(data['mean']),label='{} mm'.format(name[i]))
    ax1.scatter(data['X']-centers[i], watts_per_mm2(data['mean']))
    
    ax2.semilogy(data['X']-centers[i], watts_per_mm2(data['mean']),label='{} mm'.format(name[i]))

ax1.legend()
ax1.set_xlabel('CIMACs X Axis (mm)')
ax1.set_ylabel("Watts$/mm^{2}$")
ax2.set_xlabel('CIMACs X Axis (mm)')
ax2.set_ylabel("Log Watts$/mm^{2}$")
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
```

```python
output_notebook()
```

```python
data_df = pd.read_csv(scan_full)
```

```python
data_df['mean'] = data_df['mean']# - background
data_df['Start'] = pd.to_datetime(data_df['Start'])
data_df['End'] = pd.to_datetime(data_df['End'])
source = ColumnDataSource(data=data_df)
```

```python
TOOLTIPS = [
    ("index", "$index"),
    ("start", "@Start"),
    ("end", "@End"),
    ("exptime", "@ExpTime"),
    ("(x,y)", "(@X,@Y)"),
    ("wave", "@wave"),
    ("mean", "@mean"),
    ("median", "@median"),
]
```

```python
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar
from bokeh.palettes import Spectral6, Turbo256, Viridis256
mapper = linear_cmap(field_name='mean', palette=Viridis256 ,low=min(data_df['mean']) ,high=max(data_df['mean']))
p = figure(width=600, height=500, tooltips=TOOLTIPS)
p.scatter(x='X', y='Y', source=source, size=15, color=mapper, alpha=0.7)
p.y_range.flipped = True
color_bar = ColorBar(color_mapper=mapper['transform'], width=8,  location=(0,0),title="Current (A)")
p.xaxis.axis_label = "X Stage (mm)"
p.yaxis.axis_label = "Y Stage (mm)"
p.add_layout(color_bar, 'right')

show(p)
```
